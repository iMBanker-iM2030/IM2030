```python
nonUse = df[(df['수신_요구불예금'] == 0) & (df['수신_거치식예금'] == 0) & (df['수신_적립식예금'] == 0) & (df['수신_펀드'] == 0) & (df['수신_외화예금'] == 0) & (df['대출금액'] == 0)]
```

```python
nonUse.info
```

```python
count_nonUse = nonUse[['고객ID', '기준년월']].groupby(['고객ID']).count().reset_index()
```

```python
ids_eda = count_nonUse[(count_nonUse['기준년월']> 5) & (count_nonUse['기준년월']< 8)]['고객ID'].unique()
```

```python
ids_sample = df[df['고객ID'].isin(ids_eda)].groupby('고객ID').count()['기준년월'].reset_index().groupby('기준년월').sample(n= 3)['고객ID'].unique()
```

```python
ids_sample
```

```python
samp_min = df[df['고객ID'].isin(ids_sample)].groupby('고객ID').min('기준년월')['기준년월'].sort_index()
```

```python
samp_max = df[df['고객ID'].isin(ids_sample)].groupby('고객ID').max('기준년월')['기준년월'].sort_index()
```

```python
sample=pd.concat([samp_min, samp_max], axis = 1, keys=['최초거래','최후거래'])
```

```python
sample = sample.sort_values('최초거래').reset_index() # 5 ~ 6개월 거래
```

```python
for id in sample['고객ID']:
    print(id)
    print(df[df['고객ID']==id][['기준년월', '수신_요구불예금', '수신_거치식예금', '수신_적립식예금', '수신_펀드', '수신_외화예금', '대출금액']])
```

```python
churnIds = count_nonUse[count_nonUse['기준년월']>=12].sort_values(by='기준년월')['고객ID'].unique()
```

```python
nonUsers = df[df['고객ID'].isin(churnIds)]
```

```python
df = df[~df['고객ID'].isin(churnIds)] # 임의 고객대상 추적 탐색
```

```python
df.drop(['수신_거치식예금', '수신_적립식예금', '수신_펀드', '수신_외화예금', '대출금액'], axis=1, inplace=True)
```
